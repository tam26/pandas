```python
import pandas as pd
import numpy as np
```

```python
file = "purchase_data.csv"

raw_data = pd.read_csv(file)

raw_data.head()
```

```python
player_list = raw_data["SN"].unique()
player_count = len(player_list)
total_player_count = pd.DataFrame({"Total Players":[player_count]})
total_player_count.head()
```

```python
item_list = raw_data["Item Name"].unique()
num_purchases = len(raw_data["Purchase ID"])
total_revenue = raw_data["Price"].sum()

avg_price = round(total_revenue/num_purchases,2)

purchase_summary = pd.DataFrame({
            "Number of Unique Items": [num_purchases],
            "Average Price":["$"+str(avg_price)],
            "Number of Purchases": [num_purchases],
            "Total Revenue": ["$"+str(total_revenue)]
})

purchase_summary
```

```python
duplicate_sn_removed = raw_data.drop_duplicates("SN")
gender_count = duplicate_sn_removed["Gender"].value_counts()
gender_percent = round((gender_count/player_count)*100,2)
gender_count_df = pd.DataFrame({
            "Total Count": gender_count,
            "Percentage of Players": gender_percent
})

gender_count_df.head()
```

```python
gender_group = raw_data.groupby(["Gender"])
purchase_count = gender_group["Purchase ID"].count()
total_purchase_value = gender_group["Price"].sum()
avg_purchase_price = round(total_purchase_value/purchase_count,2)
avg_purchase_per_person = round(total_purchase_value/gender_count,2)
gender_summary = pd.DataFrame({
            "Purchase Count": purchase_count,
            "Average Purchase Price": avg_purchase_price.apply("${0:.2f}".format),
            "Total Purchase Value": total_purchase_value.apply("${0:.2f}".format),
            "Avg Total Purchase per Person":avg_purchase_per_person.apply("${0:.2f}".format)
})

gender_summary.head()
```

```python
bins = [0,9,14,19,24,29,34,39,100]
names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

age_summary = pd.cut(duplicate_sn_removed["Age"], bins, labels=names)

age_count = age_summary.value_counts()

percent_age = round((age_summary.value_counts())/(player_count)*100,2)

age_breakdown = pd.DataFrame({
            "Total Count": age_count,
            "Percentage of Players": percent_age,
})

age_breakdown_sorted = age_breakdown.sort_index(ascending=True)
age_breakdown_sorted
```

```python
raw_data["Age Group"] = pd.cut(raw_data["Age"], bins, labels=names)

total_purchase_valye_by_age = raw_data.groupby(["Age Group"]).sum()["Price"]
purchase_count_by_age = raw_data.groupby(["Age Group"]).count()["Price"]
avg_purchase_by_age = raw_data.groupby(["Age Group"]).mean()["Price"]

purchasing_analysis_by_age = pd.DataFrame({"Purchase Count": purchase_count_by_age,
         "Average Purchase Price":avg_purchase_by_age.apply("${0:.2f}".format),
         "Total Purchase Value": total_purchase_valye_by_age.apply("${0:.2f}".format),
         "Avg Total Purchase per Person":avg_purchase_by_age.apply("${0:.2f}".format)
 })

purchasing_analysis_by_age
```

```python
unique_sn_count = raw_data.groupby(["SN"]).count()["Price"]
total_purchase_per_sn = raw_data.groupby(["SN"]).sum()["Price"]
avg_purchase_per_sn = raw_data.groupby(["SN"]).mean()["Price"]

top_spenders = pd.DataFrame({"Purchase Count":unique_sn_count,
                             "Average Purchase Price":avg_purchase_per_sn.apply("${0:.2f}".format),
                             "Total Purchase Price":total_purchase_per_sn.apply("${0:.2f}".format)
    
})

top_spenders = top_spenders.sort_values("Total Purchase Price",ascending=False)
top_spenders
```

```python
item_id_one = raw_data.loc[:,["Item ID", "Item Name", "Price"]]

item_count = item_id_one.groupby(["Item ID","Item Name"]).count()["Price"]
item_price = item_id_one.groupby(["Item ID","Item Name"]).mean()["Price"]
item_total = item_id_one.groupby(["Item ID","Item Name"]).sum()["Price"]

popular_items = pd.DataFrame({"Purchase Count" : item_count,
                              "Item Price" : item_price.apply("${0:.2f}".format),
                              "Total Purchase Value" : item_total.apply("${0:.2f}".format)
})

popular_items = popular_items.sort_values("Purchase Count",ascending=False)
popular_items
```

```python
popular_items = popular_items.sort_values("Total Purchase Value",ascending=False)
popular_items
```
